string.punctuation¶
String of ASCII characters which are considered punctuation characters in the C locale: !"#$%&'()*+,-./:;<=>?@[\]^_`{|}~.

```python
import string
string.punctuation
```

```python
def remove_punctuation(text):
    punctuationfree="".join([i for i in text if i not in string.punctuation])
    return punctuationfree
```

```python
#storing the puntuation free text
data['clean_msg']= data['v2'].apply(lambda x:remove_punctuation(x))
data.head()
```

```python
data['msg_lower']=data['clean_msg'].apply(lambda x:x.lower())
```

```python
data
```

```python
import re
def tokenization(text):
    tokens = re.split('W+',text)
    return tokens
data['msg_tokenied']= data['msg_lower'].apply(lambda x: tokenization(x))
data
```

```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
```

```python
data['msg_tokenied_with_nltk']= data['msg_lower'].apply(lambda x: nltk.word_tokenize(x)) 
```

```python
data
```

```python
stopwords=nltk.corpus.stopwords.words('english')
```

```python
stopwords
```

```python
def remove_stopWord(text):
    output=[i for i in text if i not in stopwords]
    return output
```

```python
data['remove_stopword']=data['msg_tokenied'].apply(lambda x: remove_stopWord(x))
```

```python
data
```

```python
data['remove_stopword1']=data['msg_tokenied_with_nltk'].apply(lambda x: remove_stopWord(x))
```

stemming

```python
from nltk.stem.porter import PorterStemmer
```

```python
porter_stemmer= PorterStemmer()
```

```python
def stemmer(text):
    stem_text=[porter_stemmer.stem(word) for word in text]
    return stem_text
```

```python
data['stemming_text']= data['remove_stopword1'].apply(lambda x: stemmer(x))
```

```python
data
```

lemmatize

```python
from nltk.stem import WordNetLemmatizer
wordnet= WordNetLemmatizer()
```

```python
def lemmatizer(text):
    lemm_text = [wordnet.lemmatize(word) for word in text]
    return lemm_text
```

```python
data['meg_lammiter']=data['remove_stopword1'].apply(lambda x: lemmatizer(x))
```

```python
data
```

```python
data['lemmatized'] = data.meg_lammiter.apply(lambda x: ' '.join(x))
```

```python
data
```

```python
x=data["lemmatized"]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
bow=CountVectorizer()
```

```python
bow_data=bow.fit_transform(x)
```

```python
bow_data.toarray().shape
```

```python
vocab= bow.get_feature_names()
```

```python
dist=np.sum(bow_data.toarray(),axis=0)
```

```python
word_freq = pd.DataFrame({'word':vocab,'freq':dist})
```

```python
word_freq.sort_values(by='freq',ascending=False)[:10]
```
